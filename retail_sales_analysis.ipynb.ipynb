```python
plt.figure(figsize=(8,5))
sns.set_style("whitegrid")

sns.barplot(
    data=df,
    x='Category',
    y='Sales',
    estimator=sum,
    palette='Set2'
)

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

# **Insight:**
The Technology category generates the highest total sales, clearly outperforming Furniture and Office Supplies.
This indicates that customer demand is strongest for technology products, making this category the primary revenue driver for the company.

# **Business implication:**
The company should prioritize inventory availability and marketing efforts for Technology products, as any disruption in this category could significantly impact overall revenu

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x='Category',
    y='Profit',
    estimator=sum,
    palette='Set2'
)

plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()
```

# **Insight:**
The Technology category generates both the highest total sales and the highest total profit among all categories.
This indicates that technology products are not only in high demand but also maintain strong profit margins.

# ðŸŽ¯ **Business implication:**

The company should continue investing in the Technology category, as it represents the most effective balance between revenue generation and profitabilit

```python
plt.figure(figsize=(8,5))
sns.set_style("whitegrid")

sns.scatterplot(
    data=df,
    x='Discount',
    y='Profit'
)

plt.title("Relationship Between Discount and Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

# **Insight:**
The scatter plot shows a clear negative relationship between discount and profit.
As discount levels increase, profit tends to decrease, with several high-discount transactions resulting in negative profit.

# **Business implication:**
Excessive discounting is hurting profitability. The company should define a maximum discount threshold and avoid deep discounts on low-margin product

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='Discount',
    y='Profit',
    hue='Category'
)

plt.title("Discount vs Profit by Category")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

# **Insight:**
The negative impact of discounts on profit is most visible in the Technology category, where high discounts often lead to losses.
Other categories appear slightly more tolerant to discounting but still show declining profitability at higher discount levels.

# **Business implication**:
Discount strategies should be category-specific, with stricter discount limits applied to Technology product

```python
plt.figure(figsize=(8,5))
sns.set_style("whitegrid")

sns.barplot(
    data=df,
    x='Region',
    y='Sales',
    estimator=sum,
    palette='Set2'
)

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

# **Insight**:
The West region records the highest total sales among all regions, indicating stronger customer demand or higher order volumes compared to other regions.

# **Business implication:**
The company should prioritize inventory distribution and marketing efforts in the West region to sustain and further grow sales performanc

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x='Region',
    y='Profit',
    estimator=sum,
    palette='Set2'
)

plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.show()
```

# **Insight**:
Although the West region leads in sales, it also generates the highest total profit, showing that strong sales performance in this region translates into healthy profitability.

# **Business implication**:
The West region represents a high-value market and should be treated as a strategic priority for expansion and customer retentio
